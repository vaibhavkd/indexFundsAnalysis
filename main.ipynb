Importing Packages

```python
#For Getting Data
from yahoofinancials import YahooFinancials

#For Manipulating Data
import pandas as pd

#For Visualizing Data
import plotly.graph_objects as go
```

```python
#Reading csv file containing funds' info (name and symbol).
fundsDF = pd.read_csv('indexFundsInput.csv')
fundsDF
```

```python
fundsDict = {}
for index in fundsDF.index:
    fundsDict[fundsDF['Name of Fund'][index]] = fundsDF['Symbol'][index]
fundsDict
```

```python
#This function gets data for all funds and joins dataframe of all funds, and outputs data into a combined dataframe.
def get_data(fundsDict, start_date, end_date):
    main_df = pd.DataFrame()

    for fund in fundsDict:
        fundName = fund.strip()
        ticker = fundsDict[fund].strip()

        try:
            yahoo_financials = YahooFinancials(ticker)
            data = yahoo_financials.get_historical_price_data(start_date, 
                                                            end_date, 
                                                            time_interval='daily')
            
            df = pd.DataFrame(data[ticker]['prices'])
            df = df.drop('date', axis=1).set_index('formatted_date')
            df = df.drop(["high", "low", "open", "close", "volume"], axis=1)
            df.rename(columns={'adjclose': fundName}, inplace=True)

            if main_df.empty:
                main_df = df
            else:
                main_df = main_df.join(df, how='outer')       
        except:
            pass

    main_df = main_df.dropna(0)

    return main_df
```

```python
def get_one_year_rolling_returns_df(main_df):
    main_df = main_df.rolling(250).mean()
    main_df = main_df.dropna(0)
    return main_df
```

```python
def get_returns(main_df):
    returns = {}
    for col in main_df.columns:
        pv = list(main_df[col])[0]
        fv = list(main_df[col])[-1]

        r = (fv - pv)/pv * 100
        returns[col] = round(r, 2)
    return returns
```

Coefficient of Variation = ( Std Dev / Mean ) * 100  %

Standard Deviation can not be expresses in percentage, and we can not use it to compare spreads(score from the mean) of two different groups.

But Coefficient of Variance can be expresses in percentages and it can be used to compare spreads(core from the mean) of two different groups.

```python
def get_coefficient_of_variation(main_df):
    stdDev = dict(main_df.std())
    output = {}
    for col in main_df.columns:
        std_dev = stdDev[col]
        output[col] = std_dev/main_df[col].mean() * 100
    return output
```

```python
def rank_funds_by_returns(main_df):
    returns = {}
    for col in main_df.columns:
        pv = list(main_df[col])[0]
        fv = list(main_df[col])[-1]

        r = (fv - pv)/pv * 100
        returns[col] = round(r, 2)
    funds_with_high_returns = []
    sortedList = sorted(returns.items(), key=lambda item: item[1], reverse=True)
    ranking = [(lambda x: list(x))(x) for x in sortedList]
    return ranking
```

At this point we have defined all important functions

CREATING CALENDER YEAR RETURNS TABLE

```python
def create_calender_year_returns():

    output_dataframe = pd.DataFrame()

    for year in range(2021, 2017, -1):
        main_df = get_data(fundsDict, '{}-01-01'.format(year), '{}-12-31'.format(year))
        ranking = rank_funds_by_returns(main_df)
        ranking.insert(0, [year, '{} returns'.format(year)])
        column_names = ranking.pop(0)
        df = pd.DataFrame(ranking, columns=column_names)

        if main_df.empty:
            output_dataframe = df
        else:
            output_dataframe = output_dataframe.join(df, how='outer')  

        print('{} is done.'.format(year))

    return output_dataframe

calenderYearReturnsDf = create_calender_year_returns()

#Converting dataframe to csv file name 'calenderYearReturnsTable.csv'
calenderYearReturnsDf.to_csv('calenderYearReturnsTable.csv', index=True)
```

ROLLING RETURNS ANALYSIS

```python
#Getting data for all funds for this particular time period.
main_df = get_data(fundsDict, '2018-08-07', '2021-08-07')

#Creating rolling returns dataframe from main dataframe.
oneYearRRdf = get_one_year_rolling_returns_df(main_df)

#Ranking funds by their rolling returns
rankingByReturns = rank_funds_by_returns(oneYearRRdf)

#Getting dictionary for returns of all funds.
returnsDict = get_returns(main_df)

#Getting dictionary for 'coefficient of variation' of all funds.
coeffOfVarForRollingReturnsDict = get_coefficient_of_variation(oneYearRRdf)
```

Approach Explained

First we rank funds by their rolling returns, 
but we also have to consider their variance in our analysis. 
For this we will be using "coefficent of variance (%)" metric, 
because it allows us to compare variance of two different groups unlike standard deviation or pure variance.

Then we divide funds' ranking into four quartiles(top, middleTop, middleBottom, bottom). 
Then in each quartile, we rank the funds by their coefficent of variance. 
In short we used both returns and variance to rank funds, but gave priority to returns over variance.

Finally, we divides the ranking into 10 deciles and visualize each deciles using plotly.

```python
quartiles = []
l = len(rankingByReturns)
current = 0
next = 0

while next < l:
    next = current + int(l*25/100)
    if next > l:
        quartiles.append(rankingByReturns[current: l])
    else:
        quartiles.append(rankingByReturns[current: next])
        current = next

mainRanking = []
for quartile in quartiles:
    fundsCoeffOfVarDict = {}
    for fund in quartile:
        fundsCoeffOfVarDict[fund[0]] = coeffOfVarForRollingReturnsDict[fund[0]]
    sortedList = sorted(fundsCoeffOfVarDict.items(), key=lambda item: item[1], reverse=False)
    ranking = [(lambda x: list(x))(x) for x in sortedList]
    for rank in ranking:
        mainRanking.append(rank)

for rank in mainRanking:
    print(rank)
```

```python
#We are creating a dictionary that contains Rolling Returns and Coefficient Of Variation of all funds in our mainRanking.
res = {}
for rank in mainRanking:
    fund = rank[0]
    res[fund] = {'Returns': returnsDict[fund], 'Coefficient Of Variation': coeffOfVarForRollingReturnsDict[fund]}
```

```python
#Creating deciles(10% partitions of whole sample) from the 40+ funds ranking.
deciles = []
l = len(mainRanking)
current = 0
next = 0

while next < l:
    next = current + int(l/10)
    if next > l:
        deciles.append(mainRanking[current: l])
    else:
        deciles.append(mainRanking[current: next])
        current = next

#Visualization
for i, decile in enumerate(deciles):
    print('Decile {}'.format(i+1))
    fig = go.Figure()
    for rank in decile:
        fund = rank[0]
        name = fund + '[Returns:' + str(round(res[fund]['Returns'], 2)) + '%;' + ' COV(in %):' + str(round(res[fund]['Coefficient Of Variation'], 2)) + ']'
        print(name)
        fig.add_trace(
            go.Scatter(
                x=[x for x in oneYearRRdf.index],
                y=[y for y in oneYearRRdf[fund]],
                name = name
            )
        )
        
    fig.update_layout(
        title="{} Decile".format(i+1),
        xaxis_title="Time-line",
        yaxis_title="Adj Close",
        font=dict(
            family="Courier New, monospace",
            size=12,
            color="RebeccaPurple"
        )
    )

    fig.show()
    print('------------------------------------------------')
```
